```python
from sklearn import datasets, model_selection, naive_bayes
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
```

```python
print digits['data'][:15]
breast_cancer['data'][:15]
```

```python
estimator = naive_bayes.BernoulliNB()

err_digits_BernoulliNB = model_selection.cross_val_score(estimator=estimator, X=digits['data'], y=digits['target']).mean()
err_beast_cancer_BernoulliNB = model_selection.cross_val_score(estimator=estimator, X=breast_cancer['data'], y=breast_cancer['target']).mean()

print 'BernoulliNB\ndigits: {:15.2f}\nbreast_cancer: {:8.2f}'.format(err_digits_BernoulliNB, err_beast_cancer_BernoulliNB)

estimator = naive_bayes.MultinomialNB()
err_digits = model_selection.cross_val_score(estimator=estimator, X=digits['data'], y=digits['target']).mean()
err_beast_cancer = model_selection.cross_val_score(estimator=estimator, X=breast_cancer['data'], y=breast_cancer['target']).mean()
print '\nMultinominalNB\ndigits: {:15.2f}\nbreast_cancer: {:8.2f}'.format(err_digits, err_beast_cancer)

estimator = naive_bayes.GaussianNB()
err_digits = model_selection.cross_val_score(estimator=estimator, X=digits['data'], y=digits['target']).mean()
err_beast_cancer = model_selection.cross_val_score(estimator=estimator, X=breast_cancer['data'], y=breast_cancer['target']).mean()
print '\nGaussianNB\ndigits: {:15.2f}\nbreast_cancer: {:8.2f}'.format(err_digits, err_beast_cancer)
```
